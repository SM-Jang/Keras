```python
# sample data
    
X = np.random.randn(6,4,1) 
y = np.random.randn(6) 

print(X.shape) 
print(y.shape)
```

# Build the three RNN Models

```python
from keras.models import Sequential
from keras import layers

# many - to - many
# 1. Simple RNN
simpleRNN = Sequential([ 
    keras.layers.SimpleRNN(units=10, 
                      return_sequences=True, 
                      input_shape=[4,1]), 
    keras.layers.SimpleRNN(units=30),
    keras.layers.Dense(1) 
])


simpleRNN.compile(optimizer='adam', 
              loss='mse') 

simpleRNN.fit(X, y, epochs=100)
simpleRNN.summary()
```

```python
Image("./img/false.png")
```

```python
# 1. Simple RNN
# many - to - one
simpleRNN = Sequential([ 
    keras.layers.SimpleRNN(units=10, 
                      return_sequences=False, 
                      input_shape=[4,1]),
        keras.layers.Dense(1) 
])


simpleRNN.compile(optimizer='adam', 
              loss='mse') 

simpleRNN.fit(X, y, epochs=100)
simpleRNN.summary()
```

```python
# 2. GRU
GRUmodel = Sequential([ 
    keras.layers.GRU(units=10, 
                      return_sequences=True, 
                      input_shape=[4,1]), 
    keras.layers.GRU(units=30),
    keras.layers.Dense(1) 
])
GRUmodel.summary()

GRUmodel.compile(optimizer='adam', 
              loss='mse') 

GRUmodel.fit(X, y, epochs=100)
```

```python
# 3. LSTM
GRUmodel = Sequential([ 
    keras.layers.LSTM(units=10, 
                      return_sequences=True, 
                      input_shape=[4,1]), 
    keras.layers.LSTM(units=30),
    keras.layers.Dense(1) 
])
GRUmodel.summary()

GRUmodel.compile(optimizer='adam', 
              loss='mse') 

GRUmodel.fit(X, y, epochs=100)
```
